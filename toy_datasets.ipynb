```python
from sampledataset_generator import generator
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## Dataset 1: NaNが含まれるデータセット (回帰)
サンプルサイズ: 1000  
説明変数: 連続値x10、NaNを含む  
目的変数: 連続値

```python
# dataset生成
np.random.seed(seed=0)
dataset = generator.SampleDatasetGenerator(n_samples=1000)
dataset.generate()
dataset.fill_nan_randomly(rate=0.05)
```

```python
# 生成dataset
# dataset.X
# dataset.Y
display(pd.DataFrame(dataset.X).head())
```

```python
# datasetの確認 (2変数のみ)
plt.scatter(dataset.X[:,0], dataset.X[:,1], c=dataset.Y, cmap='bwr')
plt.show()
```

## Dataset 2: NaNが含まれるデータセット (分類)
サンプルサイズ: 1000  
説明変数: 連続値x10、NaNを含む  
目的変数: 離散値 (2 class)

```python
# dataset生成
np.random.seed(seed=0)
dataset = generator.SampleDatasetGenerator(n_samples=1000)
dataset.generate()
dataset.fill_nan_randomly(rate=0.05)
dataset.categoricalize()
```

```python
# 生成dataset
# dataset.X
# dataset.Y
display(pd.DataFrame(dataset.X).head())
```

```python
# datasetの確認 (2変数のみ)
plt.scatter(dataset.X[:,0], dataset.X[:,1], c=dataset.Y, cmap='bwr')
plt.show()
```

## Dataset 4: カテゴリ変数とNaNが含まれるデータセット (回帰)
